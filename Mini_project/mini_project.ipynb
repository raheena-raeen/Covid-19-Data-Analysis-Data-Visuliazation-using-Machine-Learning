```python
import pandas as pd
from matplotlib import pyplot as plt
state_wise_daily = pd.read_csv('state_wise_daily.csv')
swd = state_wise_daily.head(550)
swd
```

```python
#highlight your dataframe
swd.style.background_gradient(cmap='Reds')
```

```python
#separating values into CONFIRMED & DECEASED & RECOVERED
df_Confirmed=swd[swd['Status']=='Confirmed']
print(df_Confirmed)
df_Recovered=swd[swd['Status']=='Recovered']
print(df_Recovered)
df_Deceased=swd[swd['Status']=='Deceased']
print(df_Deceased)
```

```python
plt.plot(swd.Date,swd.MH )
plt.title('COVID 19 Analysis')
plt.xlabel('Date')
plt.ylabel('Maharashtra Analysis ')
plt.show()
```
